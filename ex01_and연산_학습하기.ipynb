```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 데이터 준비
- and 연산 데이터 생성하기


```python
# 방법1. 이중리스트 [[0,0,0],[0,1,0]...]
# 방법2. 딕셔너리 {컬럼명:[0,0,0,...],컬럼명:[0,0,0,...]}
dic = {'A':[0,0,1,1,1,0,1,0],'B':[0,1,0,1,0,0,1,1],'A and B':[0,0,0,1,0,0,1,0]}
df = pd.DataFrame(dic)
df
```

### 데이터 전처리

```python
# 문제(A,B컬럼), 답(실제적인 답, A and B) 분리
# numpy > 2차원배열[행] , 배열[행][열] , 배열[행,열]
# pandas > DF.loc[행,열] , DF.iloc[행:행, 열:열]

# X =
df.loc[:,['A','B']]
df.loc[:,'A':'B'] # 끝 값 포함
df.iloc[:,0:2] # 끝 값 포함x
X = df[['A','B']] # 컬럼 인덱싱 설정되어짐
# df[0:5] # : 행인덱싱(슬라이싱)

x = df.loc[:,'A':'B']
y = df['A and B'] # 1차원 시리즈로
# df.iloc[:2] , df.loc[:, 'A anc B']

# 크기확인
print('문제크기: ', x.shape)
print('답 크기: ', y.shape)
```

```python
# 훈련용, 테스트 분리

# x_train - 훈련용(0~5행 접근), x_test - 테스트용(6~7행 접근)
x_train = x.iloc[:6]
x_test = x.iloc[6:]
# y_train - 훈련용(0~5행 접근), y_test - 테스트용(6,7행 접근)
y_train = y.iloc[:6]
y_test = y.iloc[6:]
#훈련용 데이터 크기확인
print('데이터 크기: ',x_train.shape ,y_train.shape)
#테스트용 데이터 크기확인
print('테스트 크기: ',x_test.shape, y_test.shape)
```
